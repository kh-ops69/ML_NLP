```python
def wordtoidx(f):
  word_idx_dict = {}
  idx = 1
  for line in f:
    indiv_sentence_list = line.split(' ')
    for word in indiv_sentence_list:
      word = re.sub(',|;|!|\n|(|)|-', '', word)
      if word not in word_idx_dict:
        word_idx_dict[word] = idx
        idx += 1
  return word_idx_dict

first_dic = wordtoidx(new_e_lines)
second_dic = wordtoidx(new_f_lines)
first_dic.update(second_dic)
len(first_dic)
first_dic['the']                 # testing first_dic 
```

```python
# initialise a state transition matrix A and initial transition state P- make matrices for both robert and edgar text files
one_side = len(training_dic)

A0 = np.ones((one_side, one_side))
p0 = np.ones(one_side)

A1 = np.ones((one_side, one_side))
p1 = np.ones((one_side))

p1
```

```python
def get_matrixes(int_form, A_matrix, pi_matrix):
  for integer_lists in int_form:
    last_int = None
    for integer_token in integer_lists:                 # getting integers from the integer lists which were before converted into integer formats from sentences
      if last_int is None:                               # which means it has finished the inner for loop and now has moved on to the next integer list, or sentence in this case
        pi_matrix[integer_token] += 1              # update this word in the pi_matrix that represents initial transition matrix
      else:
        try:
          A_matrix[last_int, integer_token] += 1            # otherwise update it in normal matrix
        except IndexError:
          break
      last_int = integer_token
```

```python
get_matrixes([token for token, y in zip(train_split_list, y_train) if y==0], A0, p0)
```

```python
get_matrixes([token for token, y in zip(train_split_list, y_train) if y==1], A1, p1)
```

```python
# need probabilities instead of word to integer id's

A0 /= A0.sum(axis=1, keepdims=True)               # keepdims meaning preserve the dimensions as is, since we're taking sum we don't want original dimensions to change
p0 /= p0.sum()                              # same thing done for prior probability

A1 /= A1.sum(axis=1, keepdims=True)
p1 /= p1.sum()
```

```python
# retreiving log of matrix elements to get more accuracy
logA0 = np.log(A0)
logp0 = np.log(p0)

logA1 = np.log(A1)
logp1 = np.log(p1)

logp1[1], logp0[1]
```

```python
# finding out priors of labels from training set
count1 = sum(label==0 for label in y_train)                     # one for edgar and one for robert frost
count2 = sum(label==1 for label in y_train)
total = len(y_train)                 # we take sum of edgar labels and sum of robert labels and divide them by total to find their distributions
prior1 = count1/total
prior2 = count2/total
log_prior1 = np.log(prior1)
log_prior2 = np.log(prior2)
prior1, prior2          # distribution of priors 
```

```python
class MarkovModel:
  def __init__(self, log_transition_matrices, log_initials, log_priors):
    self.logpriors = log_priors
    self.logtrans = log_transition_matrices
    self.loginitials = log_initials
    self.LabelIndex = len(log_priors)                    # since log_priors is based on y_train set

  def find_probs(self, input_, label_index):
    logA = self.logtrans[label_index]                         # we are trying to get the log row (integer list of sentence) in state transition matrix by referencing label_number
    logP = self.loginitials[label_index]                   # here loginitials will just be a single dimension matrix, hence we are retreiving only one element here, not a list of elements unlike above
    
    logprob = 0
    last_idx = None

    for idx in input_:
      if last_idx is None:
        logprob += logP[idx]
      else:
        try:
          logprob += logA[last_idx, idx]
        except IndexError:
          break
      last_idx = idx
    return logprob

  def predictions(self, inputs):
    predictions = np.zeros(len(inputs))
    for i, single_input in enumerate(inputs):
      posteriors = [self.find_probs(single_input, c)+ self.logpriors[c] \
                    for c in range(self.LabelIndex)]
      pred = np.argmax(posteriors)
      predictions[i] = pred                   # inputs represents the entire text given. predictions initialises an array for storing posterior predictions (i.e. confidence that the given text belongs to a certain class)
                                     # then we can input single lines individually from the text input and find their posterior probabilities using for loop and the formula above
                                     # after which it will return the maximum probability from the list of posterior probabilities received
                                     # we can store these max predictions further inside the predictions array
                                     # argmax simply returns the max prediction from the list of predictions which says which class i.e. robert or edgar the single line belongs to
    return predictions
```

```python
clf = MarkovModel([logA0, logA1], [logp0, logp1], [log_prior1, log_prior2])
train_pred_model = clf.predictions(train_split_list)
print(f"Train acc: {np.mean(train_pred_model == y_train)}")
```

```python
test_pred_model = clf.predictions(test_split_list)
print(f"Test acc: {np.mean(test_pred_model == y_test)}")
```

```python
f1_score(y_train, train_pred_model), f1_score(y_test, test_pred_model)
```
