```python
word2vec = KeyedVectors.load_word2vec_format('GoogleNews-vectors-negative300.bin.gz',
  binary=True)
```

```python
def similar_word(w1, w2, w3):
  word = word2vec.most_similar(positive=[w1, w2], negative=[w3])
  print("%s - %s = %s - %s" % (w1, w2, word[0][0], w3))            #word[0] to get top most word from results and [0] again to exclude its score from findings
```

```python
similar_word('woman', 'weak', 'man')
similar_word('jesus', 'christian', 'allah')        # wrong result
```

```python
def similar_collection(w1):
  results = word2vec.most_similar(positive=[w1])
  print(f"words similar to {w1}")
  for word,i in results:
    print(word)
```

```python
similar_collection('kashmir')
```
